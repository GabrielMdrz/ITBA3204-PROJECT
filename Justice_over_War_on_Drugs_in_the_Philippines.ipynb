## **War on Drugs in the Philippines**

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import csv
```

## **Uploading the Datasets**

```python
PHData = 'https://raw.githubusercontent.com/GabrielMdrz/ITBA3204-PROJECT/main/Dataset/PH%20data.csv'
```

## **Reading and Printing the Dataset**

```python
PH_Data=pd.read_csv('https://raw.githubusercontent.com/GabrielMdrz/ITBA3204-PROJECT/main/Dataset/PH%20data.csv')
PH_Data
```

## **Plotting the Two Types of Killing**

```python
bar_df= PH_Data['Type of killing'].value_counts().to_frame('count').reset_index()
bar_df = bar_df.sort_values('index')
bar_df.plot.bar(x='index', figsize=(12, 8))

plt.xlabel("Type of killing")
plt.ylabel("No. of killed")
plt.title("Count of Deaths")
```

```python
PH_Data.fillna(0 , inplace=True)

PH_Data
```

## **Locating "Police killing" in the dataset**

```python
Locate = PH_Data.dropna(subset=['Type of killing'])
Locate = PH_Data[PH_Data['Type of killing'].str.contains('Police killing')]

Locate
```

## **Plotting Deaths Report in every City**

```python
plt.figure(figsize=(15, 8))
sns.countplot(x="City", hue="Type of killing", data=PH_Data)
plt.title("Deaths Report in Every City")
```

## **Plotting Reports for every Police Station**

```python
rep_df = PH_Data['Police Station'].value_counts().to_frame('count').reset_index()
rep_df = rep_df.sort_values('index')
rep_df.plot.barh(x='index', figsize=(14, 8), color=('purple'))

plt.xlabel("No. of reported homicides")
plt.ylabel("Police Station")
plt.title("Count of Reports for every Police station")
```

## **Plotting Reported Homicides per Month in 2016-2017**

```python
plt.figure(figsize=(15, 5))
highest_no = PH_Data['Date'].value_counts().index
sns.countplot(data=PH_Data, x='Date', order=highest_no)
plt.xticks(rotation=75)
plt.title('Count of Reported Homicides per month in 2016-2017')
plt.xlabel('Date')
plt.ylabel('Count of Deaths')
```

## **Plotting Source of Infomation for Homicide**

```python
group = [np.count_nonzero(police_df['Police']), np.count_nonzero(news_df['News']), np.count_nonzero(cohr_df['Commission on Human Rights']), np.count_nonzero(hrg_df['Human Rights Groups']), np.count_nonzero(church_df['Church Groups']), np.count_nonzero(others_df['Others'] =='✔')]
label = ['Police', 'News', 'Commission on Human Rights', 'Human Rights Groups', 'Church Groups', 'Others']

plt.title('Sources of Information for Homicide ')
plt.pie(group,labels = label, radius=2.0, startangle=80, autopct='%.1f%%', )
plt.show()
```

## **Conclusion and Analysis**

##    The Stabile Center for Investigative Journalism collected this information from 23 different sources and visited four communities in Metro Manila to verify it. The data here covers the period from July 2016 to December 2017 in the following municipalities: Quezon City, Manila, and Caloocan. There are two types of homicides associated with the drug war: those committed by the police during drug stings and other operations, and those killed by unidentified assailants. For privacy and safety reasons, this dataset lists each individual killing but does not include names or other information that may identify the victims. The sources of information for each homicide are grouped into six categories: police records, news clips, the Philippine Commission on Human Rights (a government watchdog), human rights organizations, church groups, and others. A check mark indicates that the killing was documented by the source. Nearly 80 percent of these homicides have been recorded in some police document—a spot or incident report, a memo to internal affairs, or an entry in a paper or electronic blotter. Nearly 60 percent of the killings have been reported in the news. A small number have been recorded only by churches and human-rights groups.
